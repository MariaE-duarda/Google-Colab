```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('VALE 2022 A 09-2022.xlsx')
df
```

```python
df[['DATA_COMPRA']] = df[['DATA_COMPRA']].apply(pd.to_numeric)
```

```python
df[['DATA_CONTABIL']] = df[['DATA_CONTABIL']].apply(pd.to_numeric)
```

```python
df[['VENCIMENTO']] = df[['VENCIMENTO']].apply(pd.to_numeric)
```

```python
df[['PAGAMENTO']] = df[['PAGAMENTO']].apply(pd.to_numeric)
```

```python
df.dtypes
```

```python
df[df.HISTORICO == 'B'] 
```

```python
X = df.drop('ID_EMPRESA', axis=1)
# X.drop('DATA_CONTABIL', axis=1, inplace=True)
X.drop('ID_CLIENTE', axis=1, inplace=True)
X.drop('NOME', axis=1, inplace=True)
X.drop('NUMERO_CUPOM', axis=1, inplace=True)
X.drop('HISTORICO', axis=1, inplace=True)
X.drop('EXTORNO_BAIXA', axis=1, inplace=True)
X.drop('FORMAS_RECEBIMENTO', axis=1, inplace=True)
X.drop('PAGAMENTO', axis=1, inplace=True)
X.drop('VALOR_PAGO', axis=1, inplace=True)
X.drop('TRANSFERENCIA_CONTA', axis=1, inplace=True)
X.drop('ID_USUARIO', axis=1, inplace=True)
X.drop('NUMERO_LOJA', axis=1, inplace=True)
X.drop('NOME1', axis=1, inplace=True)
```

```python
X
```

```python
y = df['HISTORICO']
```

```python
for i, v in enumerate(y):
  if v == 'B':
    y[i] = 0
  else:
    y[i] = 1
y
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold, train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score

model = xgb.XGBClassifier(learning_rate=0.009, max_depth=6, min_child_weight=3,
                     subsample=0.15, colsample_bylevel=0.85, n_estimators=500)
model.fit(X_train, y_train)
p = model.predict_proba(X_test)
print(p)
y_test = y_test.astype(int)
taxa_acerto = roc_auc_score(y_test, p[::, 1])
print(f"A taxa de acerto é {taxa_acerto * 100:.2f}%")
```

```python
dataDF={}
cont = 0
for i in p:
  inadiplente = i[1] >= 50
  dataDF[i[0]] = 1 if inadiplente else 0
  cont += 1
print(dataDF)
```

```python
dfnew = pd.DataFrame(list(dataDF.items()), columns=['Probalidade Inadimplência', 'Inadimplência'])
dfnew
```
